## 4. Run the Predictions

````

%cd "/content/mmdetection/mmdet"
from mmdet.apis import init_detector, inference_detector, show_result_pyplot, show_result
import mmcv
import numpy as np

import cv2
# Load model
config_file = '/content/CascadeTabNet/Config/cascade_mask_rcnn_hrnetv2p_w32_20e.py'
checkpoint_file = '/content/epoch_36.pth'
model = init_detector(config_file, checkpoint_file, device='cuda:0')
!pip install Pillow
!pip install pdf2image
# !pip install python-poppler
# from pdf2image import convert_from_path, convert_from_bytes
import PIL
from PIL import Image
# images = convert_from_path('/content/Trash_WM.pdf')
# for img_by_page in images:
#   im1 = img_by_page("image.jpg")
# Test a single image 
img = "/content/ReliantEnergy_80001367780-1.jpg"
path=img
oimg = Image.open("/content/ReliantEnergy_80001367780-1.jpg")
fin_path=path.split('/')
path=fin_path[-1]
# Run Inference
result = inference_detector(model, img)

# Visualization results
final_display=show_result_pyplot(img, result,('Bordered', 'cell', 'Borderless'), score_thr=0.85,fig_size=(15, 10))
final_display=show_result(img, result,('Bordered', 'cell', 'Borderless'), score_thr=0.85,out_file="/content/result.png")
print(final_display)
bbox_result, segm_result = result
bboxes = np.vstack(bbox_result)
labels = [
    np.full(bbox.shape[0], i, dtype=np.int32)
    for i, bbox in enumerate(bbox_result)
]

labels = np.concatenate(labels)

print(bboxes)
cur_box=bbox_result[1]
box = cur_box.astype("int")
box=box.tolist()
#all_coords=box[0][0].split(' ')
print(box)

#print(all_coords)
c=0
store=[]
for i, (bbox, label) in enumerate(zip(bboxes, labels)):
    bbox_int = bbox.astype(np.int32)
    poly = [[bbox_int[0], bbox_int[1]], [bbox_int[0], bbox_int[3]],
            [bbox_int[2], bbox_int[3]], [bbox_int[2], bbox_int[1]]]
    x1=bbox_int[0]
    y1=bbox_int[1]
    x2=bbox_int[2]
    y2=bbox_int[3]

    
    img2 = oimg.crop((x1, y1, x2, y2))
    xdiff=x2-x1
    ydiff=y2-y1
    store.append((xdiff,ydiff))
    if ydiff>142 and xdiff>400:
      img2.save("/content/table"+path+str(c)+".png")
      


      c=c+1
    #Croped_image = Image[int(r[1]):int(r[1]+r[3]), int(r[0]):int(r[0]+r[2])]
store.sort()
print(store)


#cur_box=box[0].split(' ')
# print(box)
# x1,y1,x2,y2=(box[0][0], box[0][1], box[0][2], box[0][3])
#cur_box=cur_box.split(' ')
# x1,y1,x2,y2,z=cur_box
# print(cur_box)
# #print(bbox_result)
# interesting_areas=[]

# output = [[x1, y1, x2, y2]]
# print(output)
# !pip install camelot-py[cv]
# !pip install ghostscript
# import camelot
# output = [[0, 0, 400, 400]]
# interesting_areas=[]
# for x in output:
#   x1, y1, x2, y2=x
#   bbox_camelot = [
#             ",".join([str(x1), str(y1), str(x2), str(y2)])
#         ][0]  # x1,y1,x2,y2 where (x1, y1) -> left-top and (x2, y2) -> right-bottom in PDF coordinate space
#         #print(bbox_camelot)
#   interesting_areas.append(bbox_camelot)
# pdf_file="/content/Trash_WM.pdf"
# pg=3
# print('interest')
# print(interesting_areas)
# output_camelot = camelot.read_pdf(
#     filepath=pdf_file, pages=str(pg), flavor="stream", table_areas=interesting_areas
# )

# output_camelot[0].df

````
